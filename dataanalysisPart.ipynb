```python
#first step: choosing and loading a dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Example: load a dataset
df = pd.read_csv("C:\\DATASETS\\finance.csv")

df.head()
# part 1 : summary stats 
# instructions to know more about your dataset 
df.info()        # columns, datatypes, nulls
df.describe()    # mean, std, min, max, etc.
df.nunique()     # number of unique values per column
#Part 2: checking missing Data
df.isnull().sum()
df.isnull().mean() * 100  # % of missing
#print(df.columns) #how to look into the dataset columns

#Part 3: 1:simple plots
# 1 Histogram of a numeric column
df['Totals.Revenue'].hist(bins=30)
plt.title("Distribution of Totals.Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()

# 2 Bar plot for a categorical column (top 10 states)
df['State'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 States by Frequency")
plt.xlabel("State")
plt.ylabel("Count")
plt.show()

# 3 Correlation heatmap of numeric variables
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation between numeric financial variables")
plt.show()

#Part 3: 2:simple plots

# 1. Which states have the highest revenues?
top_revenue_states = df.groupby("State")["Totals.Revenue"].sum().sort_values(ascending=False).head(5)
print("Top 5 states by total revenue:\n", top_revenue_states)

# 2. Average expenditure vs revenue
avg_revenue = df["Totals.Revenue"].mean()
avg_expenditure = df["Totals.Expenditure"].mean()
print(f"\nAverage Revenue: {avg_revenue:.2f}")
print(f"Average Expenditure: {avg_expenditure:.2f}")

if avg_revenue > avg_expenditure:
    print(" On average, revenues exceed expenditures.")
else:
    print(" On average, expenditures exceed revenues.")

# 3. Yearly trend (mean revenue over time)
yearly_revenue = df.groupby("Year")["Totals.Revenue"].mean()
print("\nAverage revenue per year:\n", yearly_revenue.head())

git add eda_notebook.ipynb
git commit -m "EDA on financial dataset: summary stats, missing values, simple plots"
git push origin main

```
