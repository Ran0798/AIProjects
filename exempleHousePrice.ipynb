```python
# 📌 1. Import des librairies
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 📌 2. Charger le dataset
df = pd.read_csv("C:/Users/Benchmatrix WLL/Downloads/Maisons.csv")
print("Aperçu du dataset :")
print(df.head())

# 📌 3. Séparer features (X) et target (y)
X = df[["Surface", "Nb_pieces", "Nb_sdb", "Garage", "Jardin"]]
y = df["Prix"]

# 📌 4. Créer et entraîner le modèle
model = LinearRegression()
model.fit(X, y)

# 📌 5. Afficher les coefficients
print("\n--- Résultats du modèle ---")
for feature, coef in zip(X.columns, model.coef_):
    print(f"{feature} : {coef:.2f}")
print("Intercept :", model.intercept_)

# 📌 6. Exemple de prédiction
maison_test = [[120, 4, 2, 1, 1]]  # Surface=120, 4 pièces, 2 sdb, garage, jardin
prix_pred = model.predict(maison_test)
print("\nPrix prédit pour maison test :", prix_pred[0])

# 📌 7. Visualisation (Surface vs Prix, avec la tendance)
plt.scatter(df["Surface"], y, color="blue", label="Données réelles")
plt.xlabel("Surface (m²)")
plt.ylabel("Prix (€)")
plt.title("Prix des maisons en fonction de la surface")

# Tracer tendance sur la surface seule
X_surface = df[["Surface"]]
y_pred_surface = LinearRegression().fit(X_surface, y).predict(X_surface)
plt.plot(df["Surface"], y_pred_surface, color="red", label="Régression Surface")
plt.legend()
plt.show()
```
